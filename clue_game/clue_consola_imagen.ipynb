```python
# --- IMPORTS ---
import random
import os
from IPython.display import display, Image as IPImage

# --- Configuración de imágenes ---
carpeta_imagenes = "imagenes"  # Carpeta donde están todos los .png

personajes = ["profesor_plum", "senora_green", "coronel_mustard", "senorita_scarlet", "doctor_white"]
armas = ["candelabro", "cuchillo", "pistola", "cuerda", "llave_inglesa"]
locaciones = ["biblioteca", "cocina", "salon", "habitacion", "jardin"]

# Diccionarios con rutas de imagen
img_personajes = {p: os.path.join(carpeta_imagenes, p+".png") for p in personajes}
img_armas = {a: os.path.join(carpeta_imagenes, a+".png") for a in armas}
img_locaciones = {l: os.path.join(carpeta_imagenes, l+".png") for l in locaciones}

# --- Variables del juego ---
total_rondas = 5
puntaje_total = 0
historial = []

# --- Función para mostrar imágenes ---
def mostrar_imagenes(intento, solucion):
    print("\nImágenes del intento del usuario:")
    display(IPImage(img_personajes[intento["culpable"]]))
    display(IPImage(img_armas[intento["arma"]]))
    display(IPImage(img_locaciones[intento["lugar"]]))
    
    print("\nImágenes de la solución real:")
    display(IPImage(img_personajes[solucion["culpable"]]))
    display(IPImage(img_armas[solucion["arma"]]))
    display(IPImage(img_locaciones[solucion["lugar"]]))

# --- Juego ---
for ronda in range(1, total_rondas+1):
    print(f"\n===== RONDA {ronda} =====")
    
    culpable_real = random.choice(personajes)
    arma_real = random.choice(armas)
    lugar_real = random.choice(locaciones)
    solucion = {"culpable": culpable_real, "arma": arma_real, "lugar": lugar_real}
    
    print("Selecciona tu intento:")
    print("\nPersonajes:")
    for i,p in enumerate(personajes,1):
        print(f"{i}. {p}")
    sel_p = int(input("Número del culpable: "))
    
    print("\nArmas:")
    for i,a in enumerate(armas,1):
        print(f"{i}. {a}")
    sel_a = int(input("Número del arma: "))
    
    print("\nLugares:")
    for i,l in enumerate(locaciones,1):
        print(f"{i}. {l}")
    sel_l = int(input("Número del lugar: "))
    
    intento_usuario = {
        "culpable": personajes[sel_p-1],
        "arma": armas[sel_a-1],
        "lugar": locaciones[sel_l-1]
    }
    
    acertado = (intento_usuario == solucion)
    puntos = 3 if acertado else 1
    puntaje_total += puntos
    
    historial.append({
        "ronda": ronda,
        "intento_usuario": intento_usuario,
        "solucion": solucion,
        "acierto": acertado,
        "puntos": puntos
    })
    
    # Mostrar imágenes
    mostrar_imagenes(intento_usuario, solucion)
    
    # Mostrar resultado
    print(f"\nTu intento: {intento_usuario}")
    print(f"Solución real: {solucion}")
    print("¡Correcto!" if acertado else "¡Incorrecto!")
    print(f"Puntos obtenidos: {puntos}")

# --- Reporte final ---
print("\n===== REPORTE FINAL =====")
for r in historial:
    print(f"\nRonda {r['ronda']}:")
    print(f"Intento: {r['intento_usuario']}")
    print(f"Solución: {r['solucion']}")
    print(f"Acierto: {'Sí' if r['acierto'] else 'No'}")
    print(f"Puntos: {r['puntos']}")
print(f"\nPuntaje total: {puntaje_total}")
print("¡Gracias por jugar!")
```
